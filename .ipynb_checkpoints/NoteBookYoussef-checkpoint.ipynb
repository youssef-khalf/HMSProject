```python
import pyarrow.parquet as pq
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = '/kaggle/input/hms-harmful-brain-activity-classification/train_eegs/1000913311.parquet'

# Create an empty list to store the chunks
chunks = []

# Load the data in chunks
with pq.ParquetFile(file_path) as pq_file:
    for i in range(pq_file.num_row_groups):
        # Read the chunk
        chunk = pq_file.read_row_group(i).to_pandas()
        chunks.append(chunk)

# Concatenate the chunks into a single DataFrame
df = pd.concat(chunks)
df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the file path
file_path = '/kaggle/input/hms-harmful-brain-activity-classification/train_eegs/1000913311.parquet'

# Load the data
df = pd.read_parquet(file_path)

# Remove leading and trailing spaces from column names
df.columns = df.columns.str.strip()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of EEG Signals')
plt.show()

# Create subplots for each electrode
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(20, 20))
for i, col in enumerate(df.columns):
    ax = axes[i // 4, i % 4]
    ax.plot(df.index, df[col])
    ax.set_title(col)
    ax.set_xlabel('Timestamp')
    ax.set_ylabel('EEG Signal')
plt.tight_layout()
plt.show()
```
